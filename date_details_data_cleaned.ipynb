```python
import pandas as pd
import missingno as msno # Visualising Missing Data
# Import data from 'date_details.csv' into DataFrame.
table_name = 'date_details'
csv_file_path = 'date_details.csv'
date_details_df = pd.read_csv(csv_file_path)
# Display the DataFrame
display(date_details_df)
```

```python
# Make a copy of the DataFrame
date_details_df_original = date_details_df.copy()

print("Display the DataFrame details:\n")
print(date_details_df.info(), "\n")
display(date_details_df.describe(include='all'))
display(date_details_df.describe)
display(date_details_df)
```

```python
msno.matrix(date_details_df)
```

```python
print("Check how many missing values each column has:\n")
date_details_df_null = date_details_df.isnull()  # .isna() is the same as .isnull()
print("Number of missing values in each column:\n")
display(date_details_df_null.sum())
```

```python
print("list of unique values for 'index'")
print((date_details_df["time_period"].unique()))
print(date_details_df["time_period"].nunique())
```

```python
# filtering mask created
condition_to_include = date_details_df['time_period'].astype(str).str.contains('Evening|Morning|Midday|Late_Hours')

print("Mask created to check for everything other than the category selections that are expected: ")

time_period_exclusions = date_details_df[~condition_to_include]

display(time_period_exclusions)

display(time_period_exclusions.info())
```

```python
print("Mask created to check for only what is expected in the category selections: ")

date_details_df_filtered = date_details_df[condition_to_include]
date_details_df_filtered.loc[:, 'time_period'] = date_details_df_filtered['time_period'].astype('category')
display(date_details_df_filtered)
display(date_details_df_filtered.info())
```

```python
date_details_df_filtered_day = date_details_df_filtered['day'].unique()
display(date_details_df_filtered_day)
```
